# Projeto de um Sistema de Memória

## Dados do problema

- Capacidade total: **8 Gbits**
- Barramento de endereços: **13 bits**
- O barramento de dados transporta **2 palavras por acesso**
- O barramento de dados não pode exceder **64 bits**

---

## 1. Número de endereços

Como o barramento de endereços possui 13 linhas:

$$
N = 2^{13} = 8192
$$

Logo, a memória possui **8192 endereços**.

---

## 2. Tamanho da célula

A capacidade total é:

$$
8\;\text{Gbits}=8\times2^{30}=2^{33}\;\text{bits}
$$

Como

$$
\text{Capacidade}=\text{Número de Endereços}\times\text{Tamanho da Célula}
$$

tem-se

$$
\text{Tamanho da Célula}
=\frac{2^{33}}{2^{13}}
=2^{20}\;\text{bits}
$$

Portanto,

$$
\boxed{\text{Tamanho da Célula}=2^{20}\;\text{bits}=1\,048\,576\;\text{bits}=1\;\text{Mbit}}
$$

---

## 3. Barramento de Dados

Como o barramento de dados transporta **2 palavras** e sua largura máxima é **64 bits**:

$$
BD = 64\;\text{bits}
$$

Logo,

$$
\text{Palavra}=\frac{64}{2}=32\;\text{bits}
$$

---

## 4. Barramento de Endereços

O barramento de endereços possui:

$$
13\;\text{bits}
$$

---

## Resultado Final

| Parâmetro | Valor |
|-----------|-------|
| Capacidade | 8 Gbits |
| Barramento de Endereços | 13 bits |
| Número de Endereços | 8192 |
| Tamanho da Célula | 1 Mbit |
| Tamanho da Palavra | 32 bits |
| Barramento de Dados | 64 bits |

---

## Verificação

$$
8192\times2^{20}=2^{13}\times2^{20}=2^{33}=8\;\text{Gbits}
$$

# Projeto de um Sistema de Memória 2



- Barramento de Dados (BD): **16 bits**
- Barramento de Endereços: **menor que 14 bits**
- Capacidade total da memória: **32 Gbits**

---

## 1. Tamanho da célula

Como o barramento de dados possui 16 linhas, considera-se que cada endereço referencia uma célula de 16 bits.

$$
\boxed{\text{Célula}=16\;\text{bits}}
$$

---

## 2. Capacidade total

$$
32\;\text{Gbits}=32\times2^{30}=2^{35}\;\text{bits}
$$

---

## 3. Quantidade de endereços necessários

Cada endereço armazena 16 bits:

$$
16=2^4\;\text{bits}
$$

Logo,

$$
N=\frac{2^{35}}{2^4}=2^{31}
$$

São necessários **2³¹ endereços**.

---

## 4. Barramento de Endereços

Para endereçar $2^{31}$ posições seriam necessários **31 bits**. Entretanto, o enunciado determina que o barramento de endereços possua menos de 14 bits. Assim, adota-se o maior valor permitido:

$$
\boxed{13\;\text{bits}}
$$

---

## 5. Capacidade por chip

Com 13 linhas de endereço:

$$
2^{13}=8192\;\text{endereços}
$$

Cada endereço possui 16 bits, portanto:

$$
2^{13}\times16=2^{17}\;\text{bits}
$$

Logo,

$$
\boxed{\text{Capacidade por chip}=2^{17}\;\text{bits}=128\;\text{Kbits}}
$$

---

## 6. Quantidade de chips

$$
\frac{2^{35}}{2^{17}}=2^{18}=262\,144
$$

Logo,

$$
\boxed{262\,144\;\text{chips}}
$$

---

## Resultado Final

| Parâmetro | Valor |
|-----------|-------|
| Capacidade Total | 32 Gbits |
| Barramento de Dados | 16 bits |
| Tamanho da Célula | 16 bits |
| Barramento de Endereços | 13 bits |
| Endereços por Chip | 8192 |
| Capacidade por Chip | 128 Kbits |
| Número de Chips | 262144 |

---

## Justificativa

O barramento de dados de 16 bits define o tamanho . Para armazenar 32 Gbits seriam necessários $2^{31}$ endereços, exgindo um barramento de 31 bits. Como o enunciado limita o barramento de endereços a menos de 14 bits, utiliza-se um barramento de 13 bits por chip e a memória é organizada em múltiplos chips, totalizando 262144 chips de 128 Kbits cada.
