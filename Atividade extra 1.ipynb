Atividade 1

```python

niveis = int(input("Digite o número de níveis de quantização: "))


if niveis <= 0:
    print("Número de níveis inválido.")
elif niveis == 1:
    print("Número de bits: 1")
else:
    nbits = 0
    temp = niveis
    while temp > 1:
        temp = temp / 2
        nbits += 1


    if 2**nbits < niveis:
        nbits += 1


    print(f"Número de bits: {nbits}")

aux=0
while aux<2**nbits:
  print("\n",bin(aux)[2:])
  aux+=1
```
